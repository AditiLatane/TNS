# Project 2: CIFAR10 dataset

```python
# ======================== STEP 1: IMPORT LIBRARIES =========================
import numpy as np
import matplotlib.pyplot as plt

from tensorflow.keras.datasets import cifar10
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization
from tensorflow.keras.utils import to_categorical
```

```python
# ======================== STEP 2: LOAD INBUILT CIFAR-10 DATASET =========================
(X_train, y_train), (X_test, y_test) = cifar10.load_data()
```

```python
# ======================== STEP 3: DATA INSPECTION =========================
print("Train shape:", X_train.shape)
print("Test shape:", X_test.shape)
print("Unique classes:", np.unique(y_train))
```

```python
# ======================== STEP 4: DATA CLEANING - NORMALIZATION =========================
# Scale pixel values from 0-255 to 0-1
X_train = X_train / 255.0
X_test = X_test / 255.0
```

```python
# ======================== STEP 5: LABEL ENCODING =========================
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)
```

```python
# ======================== STEP 6: VISUALIZE SAMPLE IMAGES =========================
class_names = ['Airplane', 'Car', 'Bird', 'Cat', 'Deer', 'Dog', 'Frog', 'Horse', 'Ship', 'Truck']
plt.figure(figsize=(14,4))
for i in range(16):
    plt.subplot(2,8,i+1)
    plt.imshow(X_train[i])
    plt.title(class_names[np.argmax(y_train[i])])
    plt.axis('off')
plt.show()
```

```python
# ======================== STEP 7: BUILD CNN MODEL =========================
model = Sequential()

# Convolution Block 1 --> input layer
model.add(Conv2D(32, (3,3), activation='relu', input_shape=(32,32,3))) # no.of neurons, strid, acti fun
model.add(BatchNormalization())
model.add(MaxPooling2D((2,2)))

# Convolution Block 2 --> first hidden layer
model.add(Conv2D(64, (3,3), activation='relu'))
model.add(BatchNormalization())
model.add(MaxPooling2D((2,2)))

# Convolution Block 3 --> second hidden layer
model.add(Conv2D(128, (3,3), activation='relu'))
model.add(MaxPooling2D((2,2)))

# Conv2D(
#     no. of neurons
#     kernel_size,
#     activation=None,
#     input_shape=(height, width, channels) # Only needed for the first layer
#     filters,
#     strides=(1, 1),
#     padding='valid',
# )
```

```python
# ======================== STEP 8: FLATTEN & DENSE LAYERS =========================
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.4))
model.add(Dense(10, activation='softmax'))
```

```python
# ======================== STEP 9: MODEL SUMMARY =========================
model.summary()
```

```python
# ======================== STEP 10: COMPILE MODEL =========================
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
# ======================== STEP 11: TRAIN MODEL =========================
history = model.fit(
    X_train,
    y_train,
    epochs=10,
    batch_size=64,
    validation_split=0.2
)
```

```python
# ======================== STEP 12: PLOT ACCURACY CURVE =========================
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.xlabel("Epochs")
plt.legend()
plt.show()
```

```python
# ======================== STEP 13: MODEL EVALUATION =========================
loss, accuracy = model.evaluate(X_test, y_test)
print("Test Accuracy:", accuracy)
```

```python
# ======================== STEP 14: SINGLE IMAGE PREDICTION =========================
sample = X_test[5].reshape(1,32,32,3)
prediction = model.predict(sample)
print("Predicted Class:", class_names[np.argmax(prediction)])
```

```python
# ======================== STEP 15: DISPLAY PREDICTION =========================
plt.imshow(X_test[5])
plt.title("Predicted: " + class_names[np.argmax(prediction)])
plt.axis('off')
```

# mini proj

## 1. Load and Prepare Data

```python
import tensorflow as tf
from tensorflow.keras import layers, models

# Load Fashion-MNIST
(train_images, train_labels), (test_images, test_labels) = tf.keras.datasets.fashion_mnist.load_data()

# Normalize and Reshape
train_images = train_images.reshape((60000, 28, 28, 1)).astype('float32') / 255
test_images = test_images.reshape((10000, 28, 28, 1)).astype('float32') / 255

# Class names for reference
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat', 'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
```

## 2. Define the CNN Architecture

```python
model = models.Sequential([
    # First Convolutional Block
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),
    layers.MaxPooling2D((2, 2)),

    # Second Convolutional Block
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),

    # Third Convolutional Block
    layers.Conv2D(64, (3, 3), activation='relu'),

    # Classification Head
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')
])
```

## 3. Compile and Train

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Training for 10 epochs
history = model.fit(train_images,
                    train_labels,
                    epochs=10,
                    validation_data=(test_images, test_labels),
                    batch_size=64)
```

## 4. Evaluation and Prediction

```python
# Evaluate
test_loss, test_acc = model.evaluate(test_images, test_labels, verbose=2)
print(f'\nFinal Test Accuracy: {test_acc:.4f}')

# Predict first image
import numpy as np
prediction = model.predict(test_images[0:1])
predicted_class = class_names[np.argmax(prediction)]
actual_class = class_names[test_labels[0]]

print(f"Predicted: {predicted_class} | Actual: {actual_class}")
```
